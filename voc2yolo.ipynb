```python
import os
import random
import shutil
from pathlib import Path

# Set path to images as well as paths for output 
image_dir = Path(r"C:\Users\raiom.LAPTOP-59QT21KS\Computer Vision\22067341_OmbirRai\dataset\voc_night\voc_night\JPEGImages")
label_dir = Path(r"C:\Users\raiom.LAPTOP-59QT21KS\Computer Vision\22067341_OmbirRai\dataset\yolo_night")
output_dir = Path(r"C:\Users\raiom.LAPTOP-59QT21KS\Computer Vision\22067341_OmbirRai\dataset\yolo_night")

# Create output folders for images and corresponding labels 
for split in ["train", "val"]:
    (output_dir / f"images/{split}").mkdir(parents=True, exist_ok=True)
    (output_dir / f"labels/{split}").mkdir(parents=True, exist_ok=True)

# Split the images randomly into training and validation dataset in ratio 80:20
images = list(image_dir.glob("*.jpg"))
random.shuffle(images)
split_index = int(0.8 * len(images))
train_imgs = images[:split_index]
val_imgs = images[split_index:]

# Function to move image and corresponding label files to the designated output directories 
# based on the dataset split ("train" or "val").
# It checks if a label file exists for each image, and if so, copies both the image and label 
# to their respective split folders under 'images' and 'labels'.
def move_data(img_list, split):
    for img_path in img_list:
        img_name = img_path.stem
        label_path = label_dir / f"{img_name}.txt"

        if label_path.exists():
            shutil.copy(img_path, output_dir / f"images/{split}" / img_path.name)
            shutil.copy(label_path, output_dir / f"labels/{split}" / f"{img_name}.txt")

move_data(train_imgs, "train")
move_data(val_imgs, "val")
```
